```python
from transformers import AutoTokenizer

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Input text
text = "unhappiness playing"

# Perform subword tokenization
tokens = tokenizer.tokenize(text)

# Display tokens
print("Subword Tokens:", tokens)
```

```python
# List of common stop words
stop_words = ["is", "am", "are", "the", "a", "an", "and", "to", "in", "of"]

# Input text
text = "Tokenization is an important step in natural language processing"

# Tokenization using white space
tokens = text.split()

# Remove stop words
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# Display output
print("Tokens after stop word removal:", filtered_tokens)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Download required tokenizer models
nltk.download('punkt')

# Input text
text = "Natural Language Processing is interesting. Tokenization is the first step."

# Sentence tokenization
sentences = sent_tokenize(text)

# Word tokenization
words = word_tokenize(text)

# Display output
print("Sentences:")
print(sentences)

print("\nWords:")
print(words)
```
