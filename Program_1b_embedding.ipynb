```python
# Input text
words = ["NLP", "is", "fun"]

# Create vocabulary
vocab = list(set(words))

# One-hot encoding
one_hot_vectors = {}

for word in vocab:
    vector = [0] * len(vocab)
    index = vocab.index(word)
    vector[index] = 1
    one_hot_vectors[word] = vector

# Display output
for word, vector in one_hot_vectors.items():
    print(word, ":", vector)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Input sentences
sentences = [
    "NLP is fun",
    "NLP is easy"
]

# Create Bag of Words model
vectorizer = CountVectorizer()

# Transform text into BoW vectors
bow = vectorizer.fit_transform(sentences)

# Display vocabulary and vectors
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Bag of Words Matrix:\n", bow.toarray())
```

```python
from gensim.models import Word2Vec

# Input sentences
sentences = [
    ["nlp", "is", "fun"],
    ["nlp", "is", "easy"],
    ["machine", "learning", "is", "powerful"]
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=50, window=3, min_count=1)

# Get vector for a word
vector = model.wv["nlp"]

# Display vector
print("Vector for 'nlp':")
print(vector)
```

```python
import gensim.downloader as api

# Load pre-trained GloVe model
glove_model = api.load("glove-wiki-gigaword-50")

# Get embedding for a word
vector = glove_model["nlp"]

# Display vector
print("GloVe vector for 'nlp':")
print(vector)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Input text
text = "Natural Language Processing is interesting"

# Tokenize input
inputs = tokenizer(text, return_tensors="pt")

# Generate embeddings
with torch.no_grad():
    outputs = model(**inputs)

# Get sentence embedding (CLS token)
sentence_embedding = outputs.last_hidden_state[:, 0, :]

print("BERT Embedding:")
print(sentence_embedding)
```
